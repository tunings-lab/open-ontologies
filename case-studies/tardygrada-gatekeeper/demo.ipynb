# tardygrada-gatekeeper: a proof-carrying-action gatekeeper, exhaustively verified

An action is dispatched only if it carries a certificate the gatekeeper can check against a safety spec. Because the multi-agent system is bounded, we verify safety **completely** (not sampled): over the whole reachable state space the gate admits an action iff the result is safe.

**Runtime -> Run all.** Pure Python standard library, no installs.
